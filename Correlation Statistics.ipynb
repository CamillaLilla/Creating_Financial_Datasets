# Correlation Statistics

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib as mpl
```

```python
initial_data = yf.download(" SPY ^VIX", start="2021-03-01",end="2022-03-01")
```

```python
initial_data['Adj Close']
data =  initial_data['Adj Close']
data.info()
```

```python
Corr_data = data[['SPY', '^VIX']].dropna()
Corr_data.tail()
```

```python
Corr_data.plot(subplots=True, figsize=(10, 6))
```

```python
Corr_data.loc['2020-01-01':].plot(secondary_y='^VIX', figsize=(10, 6));
```
