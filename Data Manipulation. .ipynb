# Getting data with different time frequencies:

```python
import yfinance as yf 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = yf.download(
        tickers=['BTC-USD'], period="5d",interval="1m")

#use period instead of start/end. Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max. Default is '1mo'.

# fetch data by interval (including intraday if period < 60 days)
# valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo, default is '1d'.

data.Close.plot()
plt.show()
```

```python
data.head(10)
```

```python
data.resample('30T').mean()

#Use the resample function to resample the data into 30 minute time periods. 
#30T” = time of 30 minutes 
#.mean() calculates the average of all the values from the 30 minute time period.
```

# Changes over time:

```python
data = yf.download('AAPL','2020-01-01','2022-01-01')
data
```

```python
data.diff().head()

#calculates the absolute differences using pandas library.
```

```python
data.diff().describe().round(2)

#provides data description
```

# Derive the percentage changes or percentage returns:

```python
data.pct_change().round(4).head()
```

```python
data.pct_change().mean().plot(kind='bar', figsize=(10, 6));

#plots the histogram of the percentage difference.
```

# Log returns: alternative to percentage returns. 

```python
rets = np.log(data / data.shift(1))
rets.head().round(4)
```

```python
rets.cumsum().apply(np.exp).plot(figsize=(10, 6));
```
