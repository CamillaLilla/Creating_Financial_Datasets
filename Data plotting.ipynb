# Getting data of different types of assets:

```python
data = yf.download(['BTC-USD', 'EURUSD=X', 'CL=F', '^TNX' ],'2020-01-01','2020-12-31')

#BTC-USD =  Bitcoin in USD
#EURUSD=X = Currency pair
#CL=F = Crude oil Commodity 
#^TNX = US Treasury bond (10 years)

data
```

# Getting data with different time frequencies: 

```python
data = yf.download(
        tickers=['BTC-USD'], period="5d",interval="1m")

#use period instead of start/end. Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max. Default is '1mo'.

# fetch data by interval (including intraday if period < 60 days)
# valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo, default is '1d'.

data.Close.plot()
plt.show()
```
